```python
#Inisiasi library
import pandas as pd
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By
from time import sleep
from selenium.webdriver.common.action_chains import ActionChains
```

```python
#Menginstall driver browser untuk Selenium

browser = webdriver.Chrome(ChromeDriverManager().install())
browser.maximize_window()
#Menggunakan Selenium untuk membuka website
browser.get('https://www.bizhare.id/investasi')
while True:
    try:
        sleep(5)
        content2 = browser.find_element_by_class_name('invest-more__button')
        content2.click()
    except:
        break
SCROLL_PAUSE_TIME = 0.5

# Get scroll height
last_height = browser.execute_script("return document.body.scrollHeight")

while True:
    # Scroll down to bottom
    browser.execute_script("window.scrollTo(0, document.body.scrollHeight);")

    # Wait to load page
    sleep(SCROLL_PAUSE_TIME)

    # Calculate new scroll height and compare with last scroll height
    new_height = browser.execute_script("return document.body.scrollHeight")
    if new_height == last_height:
        break
    last_height = new_height
```

```python
containers = browser.find_elements(By.CLASS_NAME, 'invest-items')
title = []
category = []
investtype = []
nilaibisnis = []
jumlahinvestor = []
danaterkumpul = []
hari = []
sahamtersedia = []
```

```python
for i,container in enumerate(containers):
    if i%10==0:
        print(i)
    element = container.find_element(By.CLASS_NAME, 'invest-items-bottom__title')
    actions = ActionChains(browser)
    actions.move_to_element(element).perform()
    title.append(container.find_element(By.CLASS_NAME, 'invest-items-bottom__title').text)
    category.append(container.find_element(By.CLASS_NAME, 'invest-items-bottom__sub-title').text.split('\n')[0])
    investtype.append(container.find_element(By.CLASS_NAME, 'invest-items-bottom__sub-title').text.split('\n')[1])
    nilaibisnis.append(container.find_element(By.CLASS_NAME, 'invest-items-bottom-text__left').text)
    jumlahinvestor.append(container.find_element(By.CLASS_NAME, 'invest-items-bottom-text__right').text)
    danaterkumpul.append(container.find_element(By.CLASS_NAME, 'invest-items-bottom__progress-amount').text)
    if danaterkumpul[i]=="100%":
        sahamtersedia.append(container.find_element(By.CLASS_NAME, 'invest-items-bottom__slot--green').text)
    else:
        
        sahamtersedia.append(container.find_element(By.CLASS_NAME, 'invest-items-bottom__slot--red').text)
    try:
        hari.append(container.find_element(By.CLASS_NAME, 'invest-items-top__time--done').text)
    except:
        hari.append(container.find_element(By.CLASS_NAME, 'invest-items-top__time--ongoing').text)
        pass
#Menutup browser
browser.close()
```

```python
data = {
        "Title": title,
        "Category": category,
        "InvestType": investtype,
        "NilaiBisnis": nilaibisnis,
        "JumlahInvestor": jumlahinvestor,
        "DanaTerkumpul": danaterkumpul,
        "Hari": hari,
        "SahamTersedia": sahamtersedia
    }

df = pd.DataFrame(data)
df.to_csv('D:/Sagara/bizhare.csv', index=False)
print('done')
```
